```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotnine as pn
from plotnine import data
import warnings
warnings.filterwarnings('ignore')
```

```python
mpg = pn.data.mpg
mpg.head(20)
```

```python
mpg['class'].value_counts()
```

```python
sns.catplot(x='class', kind='count', data=mpg)
plt.xlabel('Type of Car')
plt.xticks(rotation=45)
plt.show()
```

tipe mobil suv menjadi tipe mobil paling banyak

```python
x=sns.countplot(mpg['class'], hue=mpg['drv'])
x.bar_label(x.containers[0])
x.bar_label(x.containers[1])
x.bar_label(x.containers[2])
plt.xlabel('Car Type')
plt.title("Perbandingan the type of drive train berdasarkan type of car")
plt.show()
```

```python
sns.scatterplot(data = mpg, x = "cty", y = "hwy", hue = "class")
plt.xlabel('city miles per gallon')
plt.ylabel('highway miles per gallon')
plt.title('Hubungan city miles per gallon dengan highway miles per gallon')
plt.show()
```

```python
sns.scatterplot(data = mpg, x = "displ", y = "cty", hue = "drv")
plt.xlabel('engine displacement, in litres')
plt.ylabel('city miles per gallon')
plt.title('Hubungan engine displacement dengan city miles per gallon')
plt.show()
```

```python
sns.scatterplot(data = mpg, x = "displ", y = "hwy", hue = "class")
plt.xlabel('engine displacement, in litres')
plt.ylabel('highway miles per gallon')
plt.title('Hubungan engine displacement dengan highway miles per gallon')
plt.show()
```
