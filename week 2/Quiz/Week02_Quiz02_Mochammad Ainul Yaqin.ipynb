```python
import pandas as pd

df = pd.read_csv('USairpollution_quiz2.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
a = df.sort_values(by='popul', ascending = False)
a = a[0:5]
plt.figure(figsize=(6,4))
plt.bar(a['City'], a['popul'])
plt.title('5 kota dengan populasi tertinggi')
plt.ylabel('Jumlah Populasi')
plt.xlabel('Nama Kota')
plt.show()
```

Berdasarkan grafik di atas dapat diketahui bahwa 5 kota dengan jumlah populasi tertinggi secara berturut-turut adalah Chicago, Philadelphia, Detroit, Houston, dan Baltimore

```python
b = df.sort_values(by='manu', ascending = False)
b = b[0:5]
plt.figure(figsize=(6,4))
plt.bar(b['City'], b['manu'])
plt.title('5 kota dengan jumlah manufacturer tertinggi')
plt.ylabel('Jumlah Manufacturer')
plt.xlabel('Nama Kota')
plt.show()
```

Berdasarkan grafik di atas dapat diketahui bahwa 5 kota dengan jumlah manufacturer tertinggi secara berturut-turut adalah Chicago, Philadelphia, Detroit, Cleveland, dan St. Louis

```python
c = df.sort_values(by='SO2', ascending = False)
c = c[0:5]
plt.figure(figsize=(6,4))
plt.bar(c['City'], c['SO2'])
plt.title('5 kota dengan kadar SO2 terbesar')
plt.ylabel('Jumlah kadar SO2 di udara')
plt.xlabel('Nama Kota')
plt.show()
```

Berdasarkan grafik di atas dapat diketahui bahwa 5 kota yang memiliki jumlah kadar SO2 di udara terbesar secara berturut-turut adalah Chicago, Providence, Philadelphia, Cleveland, dan Pittsburgh.

```python
plt.figure(figsize=(8,4))
plt.scatter(df['City'],df['clusters'], c = df['clusters'])
plt.title('Sebaran Cluster')
plt.ylabel('Cluster')
plt.xlabel('Nama Kota')
plt.xticks(rotation=90)
plt.show()
```
